```python
Accident_df = pd.read_csv('Accidents1114.csv')
Accident_df
```

```python
Accident_with_LSOA = Accident_df[Accident_df['LSOA_of_Accident_Location'].notnull()][['Accident_Index','LSOA_of_Accident_Location']]
Accident_with_LSOA
```

```python
Accident_without_LSOA = Accident_df[Accident_df['LSOA_of_Accident_Location'].isnull()][['Accident_Index','Local_Authority_(District)']]
Accident_without_LSOA
```

Đối với những tai nạn có mã LSOA

```python
Accident_with_LSOA['city']='NAN'
Accident_with_LSOA['county']='NAN'
Accident_with_LSOA['country_name']='NAN'
Accident_with_LSOA['region_name']='NAN'
```

```python
Accident_with_LSOA.set_index('LSOA_of_Accident_Location',inplace=True)
Accident_with_LSOA.update(LSOA.set_index('LSOA_of_Accident_Location'))
Accident_with_LSOA.reset_index(drop=True,inplace=True)
Accident_with_LSOA
```

Đối với tai nạn không có mã LSOA

```python
Accident_without_LSOA['city']='NAN'
Accident_without_LSOA['county']='NAN'
Accident_without_LSOA['country_name']='NAN'
Accident_without_LSOA['region_name']='NAN'
```

```python
Accident_without_LSOA.set_index('Local_Authority_(District)',inplace=True)
Accident_without_LSOA.update(LAD_df.set_index('Local_Authority_(District)'))
Accident_without_LSOA.reset_index(drop=True,inplace=True)
Accident_without_LSOA
```

Tạo data frame mới từ 2 data frame kể trên

```python
Accident = pd.concat([Accident_with_LSOA,Accident_without_LSOA])
Accident
```

```python
Accident['LocationID']='NULL'
```

```python
Location = Accident[['city','county','country_name','region_name','LocationID']]
Location
```

```python
Location = Location.drop_duplicates()
Location['LocationID'] = np.arange(len(Location))+1
Location
```

```python
Accident['LocationID'] = Accident.set_index(['city','county','country_name','region_name']).index.map(Location.set_index(['city','county','country_name','region_name'])['LocationID'])
```

```python
Accident = Accident[['Accident_Index','LocationID']]
Accident
```
